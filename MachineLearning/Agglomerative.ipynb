```python
# sklearn.cluster module provides us with AgglomerativeClustering class to perform clustering on the dataset.
# As an input argument, it requires a number of clusters (n_clusters), affinity which corresponds to the type of distance metric to use while creating clusters
```

```python
#Importing required libraries
 
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
 
 
#Getting the data ready
 
data = load_iris()
df = data.data
#Selecting certain features based on which clustering is done 
df = df[:,1:3]
 
 
#Creating the model
 
agg_clustering = AgglomerativeClustering(n_clusters = 3, metric = 'euclidean', linkage = 'average')
 
#predicting the labels
agg_clustering.fit(df)
labels = agg_clustering.fit_predict(df)
```

```python
#Plotting the results
 plt.figure(figsize = (8,5))
plt.scatter(df[labels == 0 , 0] , df[labels == 0 , 1] , c = 'red')
plt.scatter(df[labels == 1 , 0] , df[labels == 1 , 1] , c = 'blue')
plt.scatter(df[labels == 2 , 0] , df[labels == 2 , 1] , c = 'green')
plt.show()
```

```python
The dendrogram function has several parameters that allow you to customize the plot:

Z: The linkage matrix.

p: The p parameter for truncate_mode.

truncate_mode: Condenses the dendrogram for large datasets.

color_threshold: Colors the descendent links below a cluster node.

orientation: Direction to plot the dendrogram ('top', 'bottom', 'left', 'right').

labels: Labels for the leaf nodes.

count_sort: Sorts the order of the nodes.

distance_sort: Sorts the nodes based on distance.
```

```python
from scipy.cluster import hierarchy
#from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

temp = hierarchy.linkage(df, 'average')
# Plot the dendrogram
plt.figure(figsize=(20, 20))
hierarchy.dendrogram(temp, labels=agg_clustering.labels_,
           orientation='top',show_leaf_counts=True)
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# Plot the dendrogram
plt.figure(figsize=(20, 20))
hierarchy.dendrogram(temp,orientation='top',show_leaf_counts=True)
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# Plot the dendrogram
plt.figure(figsize=(20, 20))
hierarchy.dendrogram(temp,orientation='top',show_leaf_counts=True,above_threshold_color="green", color_threshold=.5)
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
