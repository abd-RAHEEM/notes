```python
import pandas as pd
data = pd.read_csv('E:\ML-LAB-2024-25\DataSets\Mall_customers.csv')
print(data.head())
```

```python
Features=['Annual Income (k$)','Spending Score (1-100)']
x=data[Features]
x.head()
```

```python
 # parameter setting : 
 # MinPoints: 
# If the dataset has two dimensions, use the min sample per cluster as 4.
# If the data has more than two dimensions, the min sample per cluster should be: Min_sample(MinPoints) = 2 * Data dimension

# To calculate the value of Eps, 
# we shall calculate the distance between each data point to its closest neighbor using the Nearest Neighbours. After that, we sort them and finally plot them. 
# From the plot, we identify the maximum value at the curvature of the graph. This value is our Eps.
```

```python
from sklearn.cluster import DBSCAN

# cluster the data into five clusters
dbc = DBSCAN(eps = 8, min_samples = 4).fit(x) # fitting the model

# If you want to check the clustering result of the data, you can use the command below
labels = dbc.labels_ 
labels
```

```python
# Plot the clusters
import matplotlib.pyplot as plt
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'], c=labels, cmap='rainbow')
plt.xlabel("Income") # X-axis label
plt.ylabel("Spending Score") # Y-axis label
plt.show() # showing the plot
```
